# Heatmap for feature selection

```python
one_hot_students_data = pd.get_dummies(students_data)

corr = one_hot_students_data.corr()

# Getting the Upper Triangle of the co-relation matrix
matrix = np.triu(corr)
plt.figure(figsize=(30, 30))
# using the upper triangle matrix as mask 
sns.heatmap(corr, mask=matrix, annot=True, xticklabels=True, yticklabels=True)
```

Here, we focus mostly on the correlations towards the target class. We can observe that there are several high correlations with the three different target attributes. Dropout and Graduate have similar high correlations with opposite impacts, whereas the class Enrolled of the target has rather different correlations.
To have a good amount of feature with reasonable prediction, we consider the top 4 high correlation of each target variable and correlations higher than 0.4 or lower than -0.4.

This leads to the following features:
Application mode, 
Debtor,
Tuition fees up to date,
Gender,
Scholarship holder,
Age at enrollment,
Curricular units 1st sem (approved),
Curricular units 1st sem (grade),
Curricular units 2nd sem (approved),
Curricular units 2nd sem (grade),
Curricular units 1st sem (evaluations),
Curricular units 2nd sem (evaluations)

The last two features are considered especially for the Enrolled class.
